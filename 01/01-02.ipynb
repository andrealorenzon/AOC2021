```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("input.csv", header=None)
data.columns = ['depth']
data
```

```python
deltas = (data-data.shift(1))
plt.plot(deltas)
(deltas>0).sum()
```

```python
roll = data.rolling(3).sum()
deltas_roll = (roll-roll.shift(1))
plt.plot(deltas_roll)
(deltas_roll>0).sum()
```
